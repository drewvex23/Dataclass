# Elementary Stats - Section 1

using the data:

        x = [3, 5, 6, 8, 12, 2]

        and 

        y = [3, 5, 6, 8, 12, 2, 5]

Create the functions

        mean(data_set)       -> returns the mean
        median(data_set)     -> returns the median
        variance(data_set)   -> returns the variance
        stand_dev(data_set)  -> returns the standard deviation
        elem_stats(data_set) -> returns a dictionary with the results of all of the above, as well as the minimum and maximum


```python
import math

def mean(data):
  
  amt = 0
  whole = 0

  for num in data:
      amt+=1
      whole+=num
  mean = whole/amt
  return mean    
  
def median(data):
  
  data.sort()
  if len(data)%2 == 0:
    return (data[len(data)//2-1]+data[len(data)//2])/2
  else:
    index = int(len(data)/2)
    return data[index]

def variance(data):
  amt = 0
  for num in data:
    amt += (num-mean(data))**2
  variance = amt/len(data)  
  return variance

def stand_dev(data):
  output = math.sqrt(variance(data))
  return output

def elem_stats(data):
  new_dict = {'mean': mean(data), 'median': median(data), 'variance': variance(data), 'standard deviation': stand_dev(data),'min': min(data),'max': max(data)}
  return new_dict

x = [3, 5, 6, 8, 12, 2]
y = [3, 5, 6, 8, 12, 2, 5]
# mean(x)         # x -> 6,       y -> 5.857...
# median(y)       # x -> 5.5,     y -> 5
# variance(y)     # x -> 11,      y -> 9.55...
# stand_dev(x)    # x -> 3.32..., y -> 3.09...
# elem_stats(x)   # {'mean': 6.0, 'median': 5.5, 'variance': 11.0, 
                  # 'standard deviation': 3.3166247903554, 'min': 2, 'max': 12}

print(mean(x),mean(y))
print(median(x), median(y))
print(variance(x), variance(y))
print(stand_dev(x), stand_dev(y))
print(elem_stats(x))
print(elem_stats(y))
```

## Clean the data

We ultimately want to run the comp% data through our elementary stats function.  However, there is some work that needs to be done on that data before it can be used.

Clean the comp% data.

```python
# clean the data
# .head() to show data cleaned correctly

og_df['Comp%'] = og_df['Comp%'].str.replace('%','')
og_df['Comp%'] = og_df['Comp%'].astype('float64')
```
